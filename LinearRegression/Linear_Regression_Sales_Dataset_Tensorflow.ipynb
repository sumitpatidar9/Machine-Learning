```python
import numpy as np
import pandas as pd

df = pd.read_csv('Advertising.csv')
df = pd.DataFrame(df)
df.head()

X = df.drop(columns = ['Sales'])
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33)

display(X_train.head(2))
display(X_test.head(2))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
#numpy array
```

```python
import tensorflow as tf

input_size = 3
hidden1_size = 200
hidden2_size = 100
output_size = 1


model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden1_size, input_shape=(input_size,)),
    tf.keras.layers.Dense(hidden2_size),
    tf.keras.layers.Dense(output_size)
])

optimizer = tf.keras.optimizers.SGD(learning_rate=0.001)
model.compile(optimizer= optimizer, 
              loss='mean_squared_error',
              metrics=['accuracy'])

history = model.fit(X_train_scaled, y_train, epochs= 150, batch_size=32, validation_split=0.2)

loss,accuracy = model.evaluate(X_test_scaled, y_test)
```

```python
print("Loss: ", loss, "Accuracy: ", accuracy)

y_pre = model.predict(X_test_scaled)
y_pre
y_pre = y_pre.flatten()
print(y_pre)
print(y_test.values)
```

```python
import matplotlib.pyplot as plt

X_test1 = X_test['TV'].values
X_test2 = X_test['Radio'].values
X_test3 = X_test['Newspaper'].values

plt.figure(figsize = (6,3))

plt.scatter(X_test1, y_test, color = 'r')
plt.scatter(X_test2, y_test, color = 'b')
plt.scatter(X_test3, y_test, color = 'g')



plt.plot(np.sort(X_test1), np.sort(y_pre), color = 'r')
plt.plot(np.sort(X_test2), np.sort(y_pre), color = 'b')
plt.plot(np.sort(X_test3), np.sort(y_pre), color = 'g')
```

```python
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(X_train.columns):
    axes[i].scatter(X_test[col], y_test, color='blue', label='Actual')
    axes[i].scatter(X_test[col], y_pre, color='red', label='Predicted')
    
    axes[i].set_title(f'Best Fit Line for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Sales')
    axes[i].legend()
plt.tight_layout()
plt.show()

X_train.columns
```
